```python
import pandas as pd
```

```python
import  numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"D:\dataset for ML\student.csv")
```

```python
data
```

```python
x=data.drop('Physics',axis=1)
```

```python
x
```

```python
y=data['Physics']
y
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2)
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
df=regressor.predict(x_test)
```

```python
dd=pd.DataFrame(df)
```

```python
dd
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
mae=mean_absolute_error(y_test,df)
```

```python
mse=mean_squared_error(y_test,df)
```

```python
sk=np.sqrt(mse)
```

```python
print("Mean_absolute_error: ",mae)
print("Mean_squared_error: ",mse)
print("Squareroot of MSE: ",sk)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x_test,df)
plt.title("Student perfomance")
plt.xlabel("x_test")
plt.ylabel("Predicted")
plt.show()
```
